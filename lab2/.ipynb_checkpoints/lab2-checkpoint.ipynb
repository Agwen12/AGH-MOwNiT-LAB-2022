## __zad.1__
### Metoda Gaussa-Jordana
Napisz i sprawdz funkcje rozwiazujaca układ równan liniowych n × n metoda Gaussa-
Jordana z czesciowym poszukiwaniem elementu wiodacego. Dla dziesieciu róznych rozmiarów
macierzy współczynników wiekszych niz 500 × 500 porównaj czasy działania
zaimplementowanej funkcji z czasami uzyskanymi dla wybranych funkcji bibliotecznych.

```python
def gauss_jordan_partial_pivoting(W: npt.NDArray, N: npt.NDArray) -> npt.ArrayLike:
    A = W.copy()
    B = N.copy()
    n = len(A)
    for i in range(n):
        max_row = i
        max_val = A[i, i]
        # partial pivoting
        for j in range(i + 1, n):
            if A[j, i] > max_val:
                max_val = A[j, i]
                max_row = j
        A[[i, max_row]] = A[[max_row, i]]
        B[i], B[max_row] = B[max_row], B[i]

        for j in range(n):
            if j != i:
                ratio = A[j, i] / A[i, i]
                A[j, i] = 0
                B[j] -= B[i] * ratio
                for c in range(i + 1, n):
                    A[j, c] = A[j, c] - (ratio * A[i, c])

    for i in range(n):
        B[i] /= A[i, i]
        A[i, i] = 1
    return B
```

```python
def gauss_jordan_full_pivoting(W: npt.NDArray, N: npt.NDArray) -> npt.ArrayLike:
    A = copy.deepcopy(W)
    B = copy.deepcopy(N)
    n = len(A)
    permutations = []

    for i in range(n):
        if A[i, i] == 0:
            continue
        # get indices of max element in sub-matrix
        m_row, m_col = np.unravel_index(np.absolute(A[i:, i:]).argmax(), [n - i, n - i])
        # translate indices to initial matrix
        m_row  += i
        m_col  += i

        if i != m_row: # swap rows
            A[[i, m_row]] = A[[m_row, i]]
            B[i], B[m_row] = B[m_row], B[i]

        if i != m_col: #swap columns
            A[:, [i, m_col]] = A[:, [m_col, i]]
            permutations.append((i, m_col))

        scale = A[i, i]
        A[i] /= scale
        B[i] /= scale

        for j in range(n):
            if i != j:
                B[j] -= B[i] * A[j, i]
                A[j] -= A[j, i] * A[i]
                A[j, i] = 0

    for i, j in permutations:
        B[i], B[j] = B[j], B[i]

    return B
```

```python
L = np.array([[1, 3, 2], [2, 2, 1], [3, 5, 4]], dtype=np.float64)
B = np.array([3, 1, 4], dtype=np.float64)
print(np.linalg.solve(L, B), "Numpy solve()")
print(gauss_jordan_partial_pivoting(L, B), "Gauss, partial pivoting + scaling")
print(gauss_jordan_full_pivoting(L, B), "Gauss, full pivoting + scaling")

print("\033[92m Exact WolframAlpha solution: \033[95m [-7.50  1.25  0]")
```

```python
def time_function(W, N, func):
    start = time()
    func(W, N)
    stop = time()

    return stop - start
```

```python
t_full, t_partial, t_np = [], [], []
for i in range(2, 501):
    W, N = generate_system_of_equations(i)
    if i < 501:
        t1 = time_function(W, N, gauss_jordan_full_pivoting)
        t_full.append(t1)
    # if i < 120:
    #     t2 = time_function(W, N, gauss_jordan_partial_pivoting)
    #     t_partial.append(t2)

    t3 = time_function(W, N, np.linalg.solve)
    t_np.append(t3)

plt.scatter(range(2, 501), t_full, label="full pivoting", marker=".")
# plt.plot(range(2, 501), t_partial, label="partial pivoting")
plt.scatter(range(2, 501), t_np, label="np solve()", marker=".")
plt.legend()
plt.xlabel("number of equations")
plt.ylabel("time spent [s]")
plt.show()
```

Jak widać na ww. wykresie czas potrzebny do rozwiązania układu równań przy wykorzystaniu __*metody Gaussa-Jordana*__ z __*full pivoting*__
rośnie propocjonalnie do ilości równań. Na podstawie tego wykresu trudnym jest określenie dokładnie tej proporcjanalności, ale można zauważyć, że nie jest to zależność linowa.

## __zad. 2__
### Faktoryzacja LU
Napisz i przetestuj funkcje dokonujaca faktoryzacji $A = LU$ macierzy A (bez poszukiwania
elementu wiodacego). Sprawdz poprawnosc wyniku obliczajac $\|A−LU\|$. Zadbaj
o to zeby implementacja była in-situ. Elementy macierzy $L$ to współczynniki mnozenia
umozliwiajace wyzerowanie odpowiedniego współczynnika macierzy $A$ w trakcie procesu
eliminacji.

```python
def lu_factorization(A: npt.NDArray) -> npt.NDArray:
    W = copy.deepcopy(A)
    n = len(W)
    for i in range(n):
        if W[i, i] == 0:
            continue
        for j in range(i, n):
            if j != i:
                o = W[j, i] / W[i, i]
                W[j,i+1:] -= o * W[i, i + 1:]
                W[j, i] = o

    return W
```

```python
difference = []
for i in range(10, 500):
    W, _ = generate_system_of_equations(i)
    LU = lu_factorization(W)
    lower = np.tril(LU)
    np.fill_diagonal(lower, 1)
    difference.append(np.linalg.norm((lower @ np.triu(LU)) - W, 1))
```

```python
plt.scatter(range(10, 500), difference, marker="p")
```

Jak widać na ww. wykresie wartośc $\|A - LU\|$ jest równa 0. Nieliczne odstające punktu można wytłumaczyć błędami
zaokragleń powstającymi wraz z procesem faktoryzacji

## Funkcja wczytująca listę krawędzi z pliku
__struktura pliku__:
a b R
...
- a - pierwszy węzeł krawędzi
- b - drugi węzeł krawędzi
- R - opór

### Uwaga ###
Jeśli __R__ nie jest podana w pliku, przyjmuje ona wartość 1


```python
def load_graph(path: str) -> nx.Graph:
    edges = []
    attrs = {}
    with open(path) as file:
        for line in file:
            splt = line.split(" ")
            if len(splt) == 3:
                a, b, R = splt
                a, b, R = int(a), int(b), int(R)
                edges.append((a, b))
                attrs[(a, b)] = {"R":R}
            else:
                a, b = int(splt[0]), int(splt[1])
                edges.append((a,b))
                attrs[(a, b)] = {"R":1}


    g = nx.Graph()
    g.add_edges_from(edges)
    nx.set_edge_attributes(g, attrs)
    return g
```

## Funkcja obliczająca natężenia prądu, stosująca I i II prawo Kirchhofa

Argumenty:
- graph - graf reprezentujący układ, który chcemy symulować
- s, t - wierzchołki między któremi przyłożono napięcie
- E - napięcie zasilające


Do wyznaczania cykli prostych zastosowano funkcję __*networkx.cycle_basis*__

```python
def kirchhoff(graph: nx.Graph, s: int, t: int, E: float):
    graph.add_edge(s, t, R=0)
    n = len(graph.nodes)
    m = len(graph.edges)
    resist_matrix =np.array( [[None for _ in range(n + 1)] for _ in range(n + 1)])
    currents = np.zeros(shape=(m + 1, m))
    result = np.zeros(shape=(m + 1, 1))
    edges_data = graph.edges.data()
    for i, (a, b, attr) in enumerate(edges_data):
        currents[a, i], currents[b, i] = 1, -1
        if "R" in graph[a][b]:
            resist_matrix[a, b] = (attr["R"], i, 0)
            resist_matrix[b, a] = (-attr["R"], i, 0)
        else:
            resist_matrix[a, b] = (1, i, 0)
            resist_matrix[b, a] = (-1, i, 0)

        if (a == s and b == t) or (b == s and a == t):
            resist_matrix[a, b] = (0, i, E)
            resist_matrix[b, a] = (0, i, -E)

    for i, cycle in enumerate(nx.cycle_basis(graph)):
        ee = .0
        for j, a in enumerate(cycle):
            b = cycle[(j + 1) % len(cycle)]
            op, idx, _ee = resist_matrix[a, b]
            ee += _ee
            currents[n + i, idx] = op
        result[n + i] = ee

    res = np.linalg.lstsq(currents, result, rcond=None)[0]
    directed = nx.DiGraph()
    di_edges = [(a, b) if res[i] < 0 else (b,a) for i, (a, b) in enumerate(graph.edges)]
    directed.add_edges_from(di_edges)
    # attrs = { list(directed.edges)[i]:{"I":abs(round(res[i][0],4))} for i in range(len(res))}
    # nx.set_edge_attributes(directed, attrs)
    pos = nx.kamada_kawai_layout(directed)
    plt.figure(figsize=(27,13))
    nx.draw(directed, pos, labels={node:node for node in graph.nodes()},
    edgelist=di_edges, edge_color=np.abs(res.flatten()), width =3, edge_cmap=plt.cm.plasma,)

    nx.draw_networkx_edge_labels(directed, pos,
                                 edge_labels={(list(graph.edges.data())[i][0], list(graph.edges.data())[i][1]):abs(round(res[i][0],4)) for i in range(len(res))},
                                 font_size=14
                                 )
    plt.axis('off')
    plt.show()
    # return directed, res
```

## Uwaga

Problem nadokreślonego układu został rozwiązany poprzez zastosowanie funckji __*numpy.linalg.lstsq*__.
__*lstsq()*__ oblicza wektor $x$, który w przybliżeniu rozwiązuje rówanie $Ax=B$.
Gdy $A_{n\times n}$ oraz $rank(A) = n$ to $x$ jest (w przybliżeniu do zaokrągleń) rozwiazaniem dokładnym

## __Funkcja obliczjąca natężenia prądu w grafie, stosując metodę potencjałów węzłowych__

Argumenty:
- graph - graf reprezentujący układ, który chcemy symulować
- s, t - wierzchołki między któremi przyłożono napięcie
- E - napięcie zasilające


Do wyznaczania cykli prostych zastosowano funkcję __*networkx.cycle_basis*__

```python
def get_current(graph, a, b):
    if "R" in graph[a][b] and graph[a][b]["R"] != 0:
        i = round((graph.nodes[b]["Vol"] - graph.nodes[a]["Vol"])/graph[a][b]["R"], 3)
    else:
        i = 0
    graph[a][b]["I"] = abs(i)
    return i
```

```python
def nodal_analysis(graph: nx.Graph, s, t, E):

    graph.add_edge(s, t, R=0, E=E)
    graph.nodes[s]["Vol"] = 0 # grounding
    graph.nodes[t]["Vol"] = E
    graph.nodes[s]["idx"] = graph.nodes[t]["idx"] = None

    for a, b in graph.edges:
        if "R" not in graph[a][b]:
            graph[a][b]["R"] = 1

    idx = 0
    for node in graph.nodes:
        if "R" not in graph.nodes[node]:
            graph.nodes[node]["R"] = 1
        if "idx" not in graph.nodes[node]:
            graph.nodes[node]["idx"] = idx
            idx += 1

    Y = np.zeros(shape=[idx, idx])
    I = np.zeros(shape=[idx])

    for node in graph.nodes:
        node_idx = graph.nodes[node]["idx"]
        if node_idx is not None:
            for nei in nx.neighbors(graph, node):
                nei_idx = graph.nodes[nei]["idx"]
                R = graph[node][nei]["R"]
                Y[node_idx, node_idx] += 1 / R

                if nei_idx is not None:
                    Y[node_idx, nei_idx] = -1 / R
                elif "Vol" in graph.nodes[nei]:
                    I[node_idx] = graph.nodes[nei]["Vol"] / R
    V = np.linalg.solve(Y, I)

    for node in graph.nodes:
        node_idx = graph.nodes[node]["idx"]
        if node_idx is not None:
            graph.nodes[node]["Vol"] = V[node_idx]


    directed = nx.DiGraph()
    di_edges = [(a, b) if get_current(graph, a, b) > 0 else (b, a) for (a, b) in graph.edges ]
    directed.add_edges_from(di_edges)

    edges, colors = zip(*nx.get_edge_attributes(graph, "I").items())
    pos = nx.kamada_kawai_layout(directed)
    plt.figure(figsize=(27,13))
    nx.draw(directed, pos, labels={node:node for node in graph.nodes()},
            edgelist=edges, edge_color=colors, width =3, edge_cmap=plt.cm.plasma,)
    nx.draw_networkx_edge_labels(directed, pos,
                                 edge_labels={edges[i] : colors[i] for i in range(len(edges))},
                                 font_size=14
                                 )
    plt.axis('off')
    plt.show()
```

## Uwaga
druga wersja zadania ma 2 wady:
- Layouty często przez *kamada_kawai_layout* w tym przypadku są nieczytelne
- mimo że wartościowo rozwiązania wyglądają na poprawe, to kierunek krawędzi niekoniecznie jest zgodny z rzeczywistym kierunkiem prądu
